```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dane= pd.read_excel("/content/drive/MyDrive/praca_licencjacka/Sprawozdanie_2014_2023_zagregowane.xlsx")
```

```python
dane
```

```python
column_to_plot = "liczba_proc_nk_a"

# Grupowanie danych po roku, obliczanie średniej
mean_values = dane.groupby('rok')[column_to_plot].mean()

# Tworzenie wykresu
plt.figure(figsize=(12, 6))

# Wykres dla poszczególnych powiatów
for powiat in dane['pow'].unique():
    data_powiat = dane[dane['pow'] == powiat]
    plt.plot(data_powiat['rok'], data_powiat[column_to_plot], label=powiat)

# Średnia dla wszystkich powiatów
plt.plot(mean_values.index, mean_values.values, label="Średnia", linestyle="--", color="black", linewidth=2)

# Dodanie tytułu i etykiet
plt.title(f"Spaghetti Plot - {column_to_plot}")
plt.xlabel("Rok")
plt.ylabel(f"Wartość: {column_to_plot}")
plt.xticks(dane['rok'].unique())
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Wyświetlanie wykresu
plt.tight_layout()
plt.show()
```

```python
column_to_plot = "liczba_proc_nk_a"

# Obliczanie sumy wartości dla każdego roku
sum_value_per_year = dane.groupby('rok')[column_to_plot].sum()

# Obliczanie średniej wartości dla wszystkich lat
mean_value_all_years = dane[column_to_plot].sum()/10

# Tworzenie wykresu
plt.figure(figsize=(12, 6))

# Wartości dla poszczególnych lat
plt.plot(sum_value_per_year.index, sum_value_per_year, 'o-', label=f"Suma wartości w każdym roku", color='blue')

# Dodanie średniej dla wszystkich lat
plt.axhline(y=mean_value_all_years, color='r', linestyle='--', label=f"Średnia ze wszystkich lat")

# Dodanie tytułu i etykiet
plt.title(f"Porównanie sumy wartości i średniej z wszystkich lat dla {column_to_plot}")
plt.xlabel("Rok")
plt.ylabel(f"Wartość: {column_to_plot}")
plt.xticks(sum_value_per_year.index)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.ylim(bottom=0)
# Wyświetlanie wykresu
plt.tight_layout()
plt.show()
```
