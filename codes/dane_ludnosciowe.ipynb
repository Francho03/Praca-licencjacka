```python
import pandas as pd

# 1. Wczytanie pliku CSV
df_sprawozdanie = pd.read_excel('/content/Sprawozdanie_przemoc_2014_2023_zagregowane_corrected.xlsx',
                              dtype={'keso': str, 'rok': int})

# 2. Wczytanie pliku Excel
df_ludnosciowe = pd.read_excel('/content/drive/MyDrive/praca_licencjacka/Dane Ludnościowe/dane_ludnosciowe.xlsx',
                               sheet_name='dane_ludnosciowe',
                               dtype={'kod': str, 'rok': int})

```

```python
# Ensure columns are string
df_sprawozdanie['keso'] = df_sprawozdanie['keso'].astype(str)
df_ludnosciowe['kod'] = df_ludnosciowe['kod'].astype(str)

# Extract the 4 digits from positions 3–6 (index 2–6, exclusive of 6)
df_sprawozdanie['keso_4digit'] = df_sprawozdanie['keso'].str[2:6]

# Extract first 4 chars from 'kod', only if they are exactly 4 digits
df_ludnosciowe['kod_4digit'] = df_ludnosciowe['kod'].str[:4]
df_ludnosciowe = df_ludnosciowe[df_ludnosciowe['kod_4digit'].str.match(r'^\d{4}$')]

# Merge on extracted 4-digit codes and 'rok'
df_merged = pd.merge(
    df_sprawozdanie,
    df_ludnosciowe,
    left_on=['keso_4digit', 'rok'],
    right_on=['kod_4digit', 'rok'],
    how='inner'
)


# 4. Obliczenie stosunków
df_merged['odsetek_ogol'] = (df_merged['liczba_osob_ogol'] / df_merged['liczba_ludnosci'])*100
df_merged['odsetek_kobiety'] = (df_merged['liczba_osob_kobiety'] / df_merged['liczba_ludnosci_kobiety'])*100
df_merged['odsetek_mezczyzni'] = (df_merged['liczba_osob_mezczyzni'] / df_merged['liczba_ludnosci_mezczyzni'])*100
df_merged['odsetek_dzieci'] = (df_merged['liczba_osob_dzieci'] / df_merged['liczba_dzieci'])*100

# 5. Wybór kolumn do wyświetlenia
result = df_merged[['woj', 'pow_x', 'rok', 'kod','odsetek_ogol', 'odsetek_kobiety', 'odsetek_mezczyzni', 'odsetek_dzieci', 'liczba_ludnosci','liczba_ludnosci_mezczyzni','liczba_ludnosci_kobiety',	'liczba_dzieci']]

# 6. Wyświetlenie próbki wyników
print(result.head())
```
