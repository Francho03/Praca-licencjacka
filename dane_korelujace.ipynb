```python
mieszkania_ogolem= pd.read_excel("/content/drive/MyDrive/praca_licencjacka/Dane korelujace/raw_data/mieszkania_ogolem.xlsx",sheet_name="TABLICA", header= None, skiprows=4, names=["kod","pow","liczba_mieszkan"])
```

```python
mieszkania_ogolem["rok"]=2011
mieszkania_ogolem
```

```python
mieszkania_najmniejsze= pd.read_excel("/content/drive/MyDrive/praca_licencjacka/Dane korelujace/raw_data/mieszkania_najmnijesze.xlsx",sheet_name="TABLICA", header= None, skiprows=4, names=["kod","pow","liczba_mieszkan"])
```

```python
mieszkania_najmniejsze["rok"]=2011
mieszkania_najmniejsze
```

```python
mieszkania_odsetek = pd.merge(
    mieszkania_najmniejsze,
    mieszkania_ogolem,
    on=["kod", "rok", "pow"],
    suffixes=["_najmniejsze","_ogolem"]
)
```

```python
mieszkania_odsetek["odsetek_najmiejszych_mieszkan"] = (
    mieszkania_odsetek["liczba_mieszkan_najmniejsze"] / mieszkania_odsetek["liczba_mieszkan_ogolem"]
).round(4)
```

```python
mieszkania_odsetek=mieszkania_odsetek.drop(columns=["liczba_mieszkan_najmniejsze","liczba_mieszkan_ogolem"])
mieszkania_odsetek
```
